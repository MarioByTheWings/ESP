```python
# 3004
import math
class Circulo:
  def __init__(self, x, y, r):
    self.x = x
    self.y = y
    self.r = r
  def area(self):
    area = math.pi * (self.r ** 2)
    print("Área =", area)
  def perimetro(self):
    perimetro = 2 * math.pi * self.r
    print("Perímetro =", perimetro)
  def mostrar_propiedades(self):
    print("Radio =", self.r)
    print(f"Centro = ({self.x}, {self.y})")

circulo1 = Circulo(10, 20, 5)

circulo1.area()
circulo1.perimetro()
circulo1.mostrar_propiedades()
```

```python
# 3005
class Operacion:
  def __init__(self, x, y):
    self.x = int(x)
    self.y = int(y)
  def suma(self):
    suma = self.x + self.y
    print("Suma = ", suma)
  def multiplicacion(self):
    mult = self.x * self.y
    print("Multiplicación = ", mult)
  def division(self):
    if self.y != 0:
      div = self.x / self.y
      print("División = ", div)
    else:
      print("No se puede dividir entre 0")

pack1 = Operacion(2, 5)
pack1.suma()
pack1.multiplicacion()
pack1.division()
```

```python
# 3006
class Persona:
  def __init__(self, nombre, edad, genero):
    self.nombre = nombre
    self.edad = edad
    self.genero = genero
  def presentarse(self):
    return (f"Hola, mi nombre es {self.nombre}, tengo {self.edad} y soy {self.genero}.")
  def esAdulto(self):
    return True if self.edad >= 18 else False

persona1 = Persona("Mario", 24, "hombre")
persona2 = Persona("Maria", 23, "mujer")

print(persona1.presentarse())
print(persona2.presentarse())

persona1.esAdulto()
```

```python
# 3009
class Vehiculo:
  def __init__(self, marca, velocidad = 0):
    self.marca = marca
    self.velocidad = velocidad

  def acelerar(self, v):
    self.velocidad = self.velocidad + v

  def desacelerar(self, v):
    self.velocidad = self.velocidad - v

  def mostrar_velocidad(self):
    print(self.velocidad)

class Coche(Vehiculo):
  def __init__(self, marca, velocidad = 0, bocina = "tuuut"):
    super().__init__(marca, velocidad)
    self.bocina = bocina

  def tocar_claxon(self):
    print(self.bocina)

coche1 = Coche("Toyota")
coche1.acelerar(20)
coche1.mostrar_velocidad()
coche1.tocar_claxon()
```

```python
# 3010
class Empleado:
  def __init__(self, nombre, funcion, salario, horas_trabajadas = 0):
    self.nombre = nombre
    self.funcion = funcion
    self.salario = salario
    self.horas_trabajadas = horas_trabajadas

  def trabajar(self, num_horas):
    self.horas_trabajadas += num_horas
    print("Horas trabajadas: ", self.horas_trabajadas)

  def info_sueldo(self):
    print("Salario: ", self.salario)

  def dar_aumento(self, cantidad):
    self.salario += cantidad
    print("Salario aumentado a", self.salario)

  def info_funcion(self):
    print("Función:", self.funcion)

empleado1 = Empleado("Mario", "Estudiante", 0)
empleado1.trabajar(300)
empleado1.info_sueldo()
empleado1.dar_aumento(1200)
empleado1.info_funcion()
```

```python
# 3011
import math
class Calculo_numerico:
  def __init__(self, numero):
    self.numero = numero

  def calculo_factorial(self):
    return math.factorial(self.numero)

  def lista_divisores(self):
    divisores = []
    for i in range (1, self.numero // 2 + 1):
      if (self.numero % i == 0):
        divisores.append(i)
    divisores.append(self.numero)
    return divisores

  def esPrimo(self):
    if self.numero < 2:
      print("No primo")
    else:
      for i in range (2, int(math.sqrt(self.numero)) + 1):
        if self.numero % i == 0:
          print("No primo")
        else:
          print("Primo")

  def esPar(self):
    print ("Es par") if (self.numero % 2 == 0) else ("No es par")

n = Calculo_numerico(4)
print(n.calculo_factorial())
print(n.lista_divisores())
n.esPrimo()
n.esPar()
```

```python
# 3012
class Video:
  def __init__(self, titulo, duracion, categoria):
    self.titulo = titulo
    self.duracion = duracion
    self.categoria = categoria

  def mirar_video(self):
    print("Mirando el video...")

  def detener_video(self):
    print("Video detenido")

class Audio:
  def __init__(self, titulo, artista):
    self.titulo = titulo
    self.artista = artista

  def escuchar_audio(self):
    print("Escuchando el audio...")

  def detener_audio(self):
    print("Audio detenido")

class Media(Video, Audio):
  def __init__(self, titulo, categoria, duracion, artista):
    Video.__init__(self, titulo, duracion, categoria)
    Audio.__init__(self, titulo, artista)
  def informacion(self):
    print(self.titulo, self.categoria, self.duracion, self.artista)

media1 = Media("tit", "cat", 187, "art")
media1.informacion()
```
